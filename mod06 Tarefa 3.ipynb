```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_sinac_ro_2019 = pd.read_csv('SINASC_RO_2019.csv')
df_sinac_ro_2019.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
# Criando uma nova coluna para categorizar o APGAR5
df_sinac_ro_2019.dropna(subset=['APGAR5'], inplace=True)
df_sinac_ro_2019.loc[(df_sinac_ro_2019['APGAR5'] >= 0) & (df_sinac_ro_2019['APGAR5'] <= 3), 'cat_apgar'] = 'asfixia grave'
df_sinac_ro_2019.loc[(df_sinac_ro_2019['APGAR5'] >= 4) & (df_sinac_ro_2019['APGAR5'] <= 5), 'cat_apgar'] = 'asfixia moderada'
df_sinac_ro_2019.loc[(df_sinac_ro_2019['APGAR5'] >= 6) & (df_sinac_ro_2019['APGAR5'] <= 7), 'cat_apgar'] = 'asfixia leve'
df_sinac_ro_2019.loc[(df_sinac_ro_2019['APGAR5'] >= 8) & (df_sinac_ro_2019['APGAR5'] <= 10), 'cat_apgar'] = 'normal'
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.countplot(x='APGAR5', data=df_sinac_ro_2019, hue='GESTACAO')
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
cross_tab = pd.crosstab(df_sinac_ro_2019['APGAR5'], df_sinac_ro_2019['GESTACAO'])
cross_tab
```

####  5) Da tabela cruzada do item anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
freq_cross_tab = cross_tab.div(cross_tab.sum(axis=1), axis=0)
freq_cross_tab
```

```python
freq_cross_tab.plot.bar()
```
