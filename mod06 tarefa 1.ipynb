#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_sinasc_ro_2019 = pd.read_csv('../modulo5/Tarefa1/SINASC_RO_2019.csv')
data_sinasc_ro_2019.head()
```

####  2) Monte gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.close('all')
sns.set(rc={'figure.figsize':(15,8)})
fig, axes = plt.subplots(2,3)
data = data_sinasc_ro_2019[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']]


#Grafico de Barras da variável APGAR1
sns.countplot(ax=axes[0,0], x='APGAR1', data=data_sinasc_ro_2019)
axes[0, 0].tick_params(axis='x', rotation=30)

#Grafico APGAR5
sns.countplot(ax=axes[0,1], x='APGAR5', data=data_sinasc_ro_2019)
axes[0,1].tick_params(axis='x', rotation=30)

#Grafico SEXO
sns.countplot(ax=axes[0,2], x='SEXO', data=data_sinasc_ro_2019)
axes[0,2].tick_params(axis='x', rotation=30)

#Grafico GESTACAO
sns.countplot(ax=axes[1,0], x='GESTACAO', data=data_sinasc_ro_2019)
axes[1,0].tick_params(axis='x', rotation=45)

#Grafico ESTCIVMAE
sns.countplot(ax=axes[1,1], x='ESTCIVMAE', data=data_sinasc_ro_2019)
axes[1,1].tick_params(axis='x', rotation=30)
axes[1,1].set_xlabel('Estado Civíl Mãe')


axes[1,2].axis('off')

plt.subplots_adjust(wspace=0.75, hspace=0.55)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
fig, axs = plt.subplots()

data_apgar_consultas = data_sinasc_ro_2019[['APGAR5', 'CONSULTAS']]
sns.barplot(ax=axs, x='APGAR5', y='CONSULTAS', data=data_apgar_consultas)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')

sns.set(rc={'figure.figsize':(15,8)})

fig, axes = plt.subplots(1,4)
data = data_sinasc_ro_2019[['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']]

ax1 = sns.countplot(ax=axes[0], x='SEXO', data=data)
ax2 = sns.countplot(ax=axes[1], x='GESTACAO', data=data)
_ = ax2.set_xticklabels(ax2.get_xticklabels(), rotation=90)

ax3 = sns.countplot(ax=axes[2], x='ESTCIVMAE', data=data)
_ = ax3.set_xticklabels(ax3.get_xticklabels(), rotation=30)

ax4 = sns.countplot(ax=axes[3], x='CONSULTAS', data=data)

plt.subplots_adjust(wspace=0.75, hspace=0.55)
```
